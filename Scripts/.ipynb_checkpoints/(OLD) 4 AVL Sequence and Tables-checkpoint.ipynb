### Summary AVL Table

```python
import pandas as pd
import numpy as np
# ID: 8-2440-02-002
new = time_sorted[[ "DAYSEQ",
                         "BUSSEQ",
                         "TRIPSEQ",
                         "PINGSEQ",
                        "ID",
                         "DIRECTION",
                         "CALENDAR_ID",
                          "MESSAGE_TIMESTAMP",
                          #"LOCAL_TIMESTAMP",
                          "LONGITUDE",
                          "LATITUDE"
                          ]]
#new
Start = new[new['PINGSEQ']==1].index.tolist()
Stop = []
for each in Start:
    x = each-1 
    Stop.append(x)
    
Starts = new.iloc[Start]
Stops = new.iloc[Stop]


#match em up on ID except last #
result = pd.merge(Starts, Stops, on=["DAYSEQ", "BUSSEQ", "TRIPSEQ"])
#for every trip change, take that row as start, and row after is end
result = result[["DAYSEQ", "BUSSEQ", "TRIPSEQ", "DIRECTION_x", "MESSAGE_TIMESTAMP_x","LONGITUDE_x","LATITUDE_x","MESSAGE_TIMESTAMP_y","LONGITUDE_y","LATITUDE_y"]]
#sum_table.to_csv(r"C:\Users\Cole\Desktop\Spring2022\AVLResearch\Data\ProcessedAVL_GTFS\IntermediateAVL2.csv")       

result.to_csv(r"C:\Users\Cole\Desktop\Spring2022\AVLResearch\Data\ProcessedAVL_GTFS\SummaryAVL.csv")       
```
