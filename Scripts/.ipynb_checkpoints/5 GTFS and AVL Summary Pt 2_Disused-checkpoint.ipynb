## GTFS 2nd Sum Table (day:direction:time)

```python
new = GTFS.sort_values(["SERV_DAY_start",'dir_sign_start',"departure_time_start"],
                                       axis=0,
                                       ascending=(True, True, True))

new.to_csv(r"C:\Users\Cole\Desktop\Spring2022\AVLResearch\Data\ProcessedAVL_GTFS\test.csv")
```

```python
#get counts for day/dir
thecount = new.groupby(["SERV_DAY_start", 'trip_headsign_start']).size().reset_index()
thecount 
```

```python
#grab start and finish rows
servdaygroup = new.groupby(["SERV_DAY_start", 'trip_headsign_start'])
c = servdaygroup.head(1)
b = servdaygroup.tail(1)
air_quality = pd.concat([c, b], axis=0)
x = air_quality.sort_values(['SERV_DAY_start'])
x
```

```python
f = pd.merge(x,thecount, how = 'right', on = ["SERV_DAY_start", 'trip_headsign_start'])

#yields table of each service day, direction, count combinations
f.to_csv(r"C:\Users\Cole\Desktop\Spring2022\AVLResearch\Data\ProcessedAVL_GTFS\SummaryFinalGTFS.csv")
```

## AVL DATA 2

```python
dataPath = r"C:\Users\Cole\Desktop\Spring2022\AVLResearch\Data\ProcessedAVL_GTFS\sumtableAVL.csv"
AVL = pd.read_csv(dataPath, index_col = False)
#take 1st and last trips for each day and direction
servdaygroup = AVL.groupby(["CALENDAR_ID", 'DIRECTION'])
c = servdaygroup.head(1) #first row in each day/DIR
b = servdaygroup.tail(1) #last row in each day/dir
air_quality = pd.concat([c, b], axis=0)
x = air_quality.sort_values(["CALENDAR_ID", 'DIRECTION'])

thecount = AVL.groupby(["CALENDAR_ID", 'DIRECTION']).size().reset_index()
#trips per direction per day

f = pd.merge(x,thecount, how = 'right', on = ["CALENDAR_ID", 'DIRECTION'])
table = f.sort_values(["CALENDAR_ID", 'DIRECTION',"Literal_Time_start"],
                                       axis=0,
                                       ascending=(True, True, True))
```
